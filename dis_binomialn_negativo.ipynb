#Distribuição binomial negativa

**Definição:** Seja X uma variável aleatória que contabiliza os números de tentativas até que r sucessos sejam obtidos.

**Representação:**  X~BN(r,p)



---
**Exemplo**: Qual é a probabilidade de conseguir 4 caras em 20 lançamentos de uma moeda?


p = 0.5

1-p = 0.5

r = 4

e = 0.6


n = 20



```python
#importando pacotes
from scipy.stats import nbinom #funções estatiticas
import pandas as pd
import numpy as np #numeros
import matplotlib.pyplot as plt #grafico
import seaborn as sns #grafico
import sympy as sp
```

```python
 #Gerando uma amostra aleatória de 1000 valores da distribuição binomial negativa com os parâmetros r e p

from re import X
r = 4
p = 0.5
X = nbinom.rvs(r,p,size=1000)
print(X)
```

```python
#grafico a partir da amostra
sns.histplot(X, kde=True)
```

#Função Geradora de Momentos

A formula da fGM de uma distribuição binomial negativa é dada por:


fgm -bn.png

```python
#calculando os valores da FGM para diferentes valores de t, tendo e = 0.6
p=0.5
r=4
t=0.06 # e**t = parametro da fgm
fmg = (0.5**4)/((sp.exp(-t)+0.5-1)**4) #formula da fgm
print(f'A Função Geradora de momentos é:', fmg)
```

#Esperança

- É a primeira derivada da fgm.

A esperança tem a segunte formula:


image.png

```python
p=0.5
r=4
esperanca= r/p
print(f'A esperança da Distribuição Binomial Negativa é:', esperanca)
```

# Esperança ao quadrado


- É a segunda derivada da fgm.


A esperança ao quadrado teórico:


esp2 -bn.png

```python
#calculando a esperança ao quadrado
p=0.5
r=4
esp_qua= (r*(r-p+1))/(p**2)
print(f'A esperança ao quadrado da Distribuição Binomial Negativa é:', esp_qua)
```

#Variância

A variância teórica:

var - bn.png

```python
#calculando a variância
p=0.5
r=4
var= (r*(1-p))/(p**2)
print(f'A variância da Distribuição Binomial Negativa é:', var)
```

```python
#Gerando grafico de uma distribuição binomial, com 20 variaveis
p=0.5
r=4
n= 20
distribution = nbinom(r, p)
x_dados = np.arange(0,20)
pmf_valores = distribution.pmf(x_dados)

plt.bar(x_dados, pmf_valores)
plt.xlabel('Valores')
plt.ylabel('Probabilidade dos valores')
plt.title('Distribuição Binomial Negativa')
plt.show()
```
